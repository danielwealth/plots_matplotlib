```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV file (adjust path if needed)
file_path = r"C:\Users\PC\Downloads\company_sales_data.csv"
data = pd.read_csv(file_path)

# Calculate total profit across all months
total_profit = data['total_profit'].sum()
print(f"Total Profit across all months: {total_profit}")

# Line plot of monthly total profit
plt.plot(data['month_number'], data['total_profit'], marker='o', linestyle='-')
plt.title("Monthly Total Profit")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file (adjust path if needed)
file_path = r"C:\Users\PC\Downloads\company_sales_data.csv"
data = pd.read_csv(file_path)

# Plot Bathing Soap and Face Wash on the same subplot
plt.plot(data['month_number'], data['bathingsoap'], marker='o', label='Bathing Soap', color='blue')
plt.plot(data['month_number'], data['facewash'], marker='o', label='Face Wash', color='green')

# Chart formatting
plt.title("Monthly Sales of Bathing Soap and Face Wash")
plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.legend()
plt.grid(True)
plt.show()
```
